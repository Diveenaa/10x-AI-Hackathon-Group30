```python
import requests
import json
import pandas as pd
```

```python
url = "https://v3-api.newscatcherapi.com/api/search"

page_size = 250
page = 0
query_string = {
    "q":"\"Apple\"",
    "lang":"en",
    "sort_by":"relevancy",
    "countries": "GB",
    "page": page,
    "from_": "1 day ago",
    "page_size": str(page_size),
    
}

headers = {
    "x-api-token": "z-DutMTgUreifwRbE69uNKQbDDZkNA6T"
    }

returned_hits = 0
total_hits = 999999999
response_list = []
while returned_hits < total_hits:
    print(returned_hits)
    page += 1
    query_string['page'] = str(page)
    response = requests.request("GET", url, headers=headers, params=query_string)
    response_json = response.json()
    total_hits = response_json['total_hits']
    response_list.append(pd.DataFrame(response_json['articles']))
    returned_hits += page_size
```
